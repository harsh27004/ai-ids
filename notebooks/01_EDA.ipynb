```python
#SECTION 3: DATA CLEANING
#Step1: Handling Missing Values
df.isnull().sum().sort_values(ascending=False).head(15)
```

```python
#Checking Infinite values
numeric_df = df.select_dtypes(include=[np.number])
np.isinf(numeric_df).sum().sort_values(ascending=False).head(15)
```

```python
#Non-numeric columns
df.select_dtypes(exclude=[np.number]).columns
```

```python
#replacing infinite values with NaN
df.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
#Rechecking missing values
df.isnull().sum().sort_values(ascending=False).head(10)
```

```python
#Handling NaN values
df.fillna(df.median(numeric_only=True), inplace=True)
```

```python
df.isnull().sum().max()
```

```python
#Step 2: Remove Duplicate Flows
# Check duplicates
df.duplicated().sum()
```

```python
# Remove duplicates
df = df.drop_duplicates()
df.duplicated().sum()
```

```python
#Step 3: Fix data type issues
df['Timestamp'] = pd.to_datetime(df['Timestamp'], errors='coerce')
df.dtypes
```

```python
#Step 4: Remove constant / near-zero variance features
from sklearn.feature_selection import VarianceThreshold

selector = VarianceThreshold(threshold=0.0)
selector.fit(df.select_dtypes(include=np.number))

low_variance_features = df.select_dtypes(include=np.number).columns[~selector.get_support()]
low_variance_features
```

```python
#Step 5: Handle invalid values
df = df[df['Flow Duration'] > 0]

print("Remaining zero durations:", (df['Flow Duration'] == 0).sum())
```
