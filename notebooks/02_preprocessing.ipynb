```python
import pandas as pd

file = "../data/raw/cicids2018/Wednesday-14-02-2018_TrafficForML_CICFlowMeter.csv"

df = pd.read_csv(file, low_memory=False)

print(df.shape)
print(df["Label"].value_counts())
```

```python
df["BinaryLabel"] = df["Label"].apply(lambda x: 0 if x == "Benign" else 1)
print(df["BinaryLabel"].value_counts())
```

```python
df = df.drop(columns=["Timestamp", "Label"])
print(df.dtypes)
```

```python
import numpy as np
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.fillna(0, inplace=True)
print("Cleaned")
```

```python
df.to_csv("../data/processed/cicids_binary_clean.csv", index=False)
print("Saved successfully")
```
